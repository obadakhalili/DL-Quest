```python
import numpy as np
from common import create_transition_matrix, draw_transition_matrix, word_generator
```

```python
names = open("./names.txt").read().splitlines()
```

```python
empty_char_token = "."
vocab = [empty_char_token] + sorted(set("".join(names)))
vocab_size = len(vocab)
```

```python
N = create_transition_matrix(names, vocab, empty_char_token)
draw_transition_matrix(N, vocab)
```

```python
P = N / N.sum(axis=1, keepdims=True)


@word_generator(block_size=1, empty_char=empty_char_token, max_size=float("inf"))
def generate_word(prev_block: str):
  return np.random.choice(a=vocab, p=P[vocab.index(prev_block)])
```

```python
generated_names = []
for _ in range(32000):
  generated_names.append(generate_word())

generated_names[:100]
```

```python
draw_transition_matrix(
  create_transition_matrix(generated_names, vocab, empty_char_token), vocab
)
```
